```python
x=data.iloc[:,0:6].values
print(x)
y=data.iloc[:,-1].values
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(
    x,y, test_size=0.3, random_state=3)

```

```python
print(x_train[0:5])
```

```python
print(y_train[0:5])
```

```python
print(x_test[0:5])
```

```python
print(y_test[0:5])
```

```python
#proses training dengan backpropagation
import tensorflow as tf


DESIRED_ACCURACY = 0.99

class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs={}):
        if(logs.get('accuracy') >= DESIRED_ACCURACY):
            print("\nReached %2.2f%% accuracy, so stopping training!!" %(DESIRED_ACCURACY*100))
            self.model.stop_training = True
callbacks = myCallback()

model=tf.keras.models.Sequential() #model arsitektur
model.add(tf.keras.layers.Dense(units=6, activation='relu')) #input layer ada 6
model.add(tf.keras.layers.Dense(units=14, activation='relu')) # 14 neuron hidden layer
model.add(tf.keras.layers.Dense(units=3, activation='softmax')) #output layer , kalau untuk 1 output pakai sigmoid
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) 

history=model.fit(
    x_train, y_train, 
    epochs=2000, 
    batch_size=50, 
    validation_data=(x_test, y_test), 
    verbose=1, 
    callbacks=[callbacks]) #training
model.evaluate(x_test, y_test, batch_size=50,callbacks=[callbacks]) #evaluasi
```

```python
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', loss)
print('Test accuracy:', accuracy)
```

```python
import matplotlib.pyplot as plt
acc = history.history['accuracy']
loss = history.history['loss']
val=history.history['val_accuracy']
val_loss = history.history['val_loss']
epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val, 'b', label='Validation accuracy')

plt.title('Training accuracy')
plt.legend(loc=0)
plt.figure()
plt.show()
```

```python
history_df = pd.DataFrame(history.history)
print(history_df['val_accuracy'].max())
print(history_df['accuracy'].max())
```
